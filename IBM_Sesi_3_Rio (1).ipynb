```python
from langchain_community.llms import Replicate
from google.colab import userdata
import os
import pandas as pd

data = pd.read_csv('/content/student_depression_dataset.csv')
api_token = userdata.get('api_token')
os.environ['REPLICATE_API_TOKEN'] = api_token

parameters = {
    "top_k": 5,
    "top_p": 1.0,
    "max_tokens": 512,
    "temperature": 0.7,
    "repetition_penalty": 1.0,
    "stopping_criteria": None,
    "stopping_sequence": None
}


llm = Replicate(
    model="ibm-granite/granite-3.3-8b-instruct",
    input = parameters
)
```

```python
!pip install langchain langchain-experimental openai


from langchain_experimental.agents.agent_toolkits.pandas.base import create_pandas_dataframe_agent
from langchain.chat_models import ChatOpenAI
import pandas as pd

# Buat agent
agent = create_pandas_dataframe_agent(
    llm=llm,
    df=data,
    verbose=True,
    allow_dangerous_code=True
)
```

```python
agent.invoke("Summarize this: High financial stress correlates with depression. Extreme sleep hours are common in depressed students. Depression does not directly reduce GPA, but long-term effects may exist. Give a short summary and policy recommendations.")

#AI agents often experience access overload, run out of memory.  hufttt
```

```python
#Exploratory Data Analysis (EDA)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set tampilan grafik lebih menarik
sns.set(style="whitegrid")
```

```python
# ✅ Step 0: Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# ✅ Step 1: Depression Distribution
plt.figure(figsize=(5,3))
sns.countplot(x='Depression', data=data, palette='Reds')
plt.title('Depression Level Distribution')
plt.xlabel('Depression Level')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()

# ✅ Step 2: CGPA Distribution
plt.figure(figsize=(5,3))
sns.histplot(data['CGPA'], kde=True, color='blue')
plt.title('Student CGPA Distribution')
plt.xlabel('CGPA')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()

# ✅ Step 3: CGPA vs Depression
plt.figure(figsize=(5,3))
sns.boxplot(x='Depression', y='CGPA', data=data, palette='coolwarm')
plt.title('CGPA by Depression Level')
plt.xlabel('Depression Level')
plt.ylabel('CGPA')
plt.tight_layout()
plt.show()

# ✅ Step 4: Correlation Heatmap
# ✅ Step 4: Correlation Heatmap
data_encoded = data.copy()
for col in data_encoded.columns:
    if data_encoded[col].dtype == 'object':
        data_encoded[col] = data_encoded[col].astype('category').cat.codes

plt.figure(figsize=(10,7))  # lebih besar dari sebelumnya
sns.heatmap(data_encoded.corr(),
            annot=True,
            cmap='coolwarm',
            fmt=".2f",
            annot_kws={"size": 8})
plt.title("Correlation Matrix")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()


# ✅ Step 5: Sleep Duration vs Depression
plt.figure(figsize=(5,3))
sns.boxplot(x='Depression', y='Sleep Duration', data=data, palette='Purples')
plt.title('Sleep Duration by Depression Level')
plt.xlabel('Depression Level')
plt.ylabel('Sleep Duration (hours)')
plt.tight_layout()
plt.show()

# ✅ Step 6: Financial Stress vs Depression
plt.figure(figsize=(5,3))
sns.countplot(x='Financial Stress', hue='Depression', data=data, palette='Set2')
plt.title('Financial Stress vs Depression')
plt.xlabel('Financial Stress')
plt.ylabel('Number of Students')
plt.legend(title='Depression Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
